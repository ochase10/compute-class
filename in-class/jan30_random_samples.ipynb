```python
import numpy as np
from matplotlib import pyplot as pl
```

```python
rng = np.random.default_rng()
samp = rng.random(int(1e7))

samp = np.sqrt(samp) / (1-np.sqrt(samp))
```

```python
pl.figure(figsize=(8,6))
bins = np.logspace(-2,6,25)
counts, _ = np.histogram(samp, bins=bins)
plot_bins = np.array([bins[i]*(bins[1]/bins[0])**0.5 for i in range(len(bins)-1)])
pl.scatter(plot_bins, counts)
pl.plot(plot_bins, 7e6*(2*np.pi*plot_bins*(1+plot_bins)**3)**-1*4*np.pi*plot_bins**3)
pl.xscale("log")
pl.yscale("log")
pl.show()
```
